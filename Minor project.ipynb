<h3> Adding a new column "Total score" and computing it by adding the test scores</h3>

```python
sum_column = df_data_1["math score"] + df_data_1["reading score"] + df_data_1["writing score"]
df_data_1["Total score"] = sum_column
```

```python
df_data_1.head()
```

<h3> Adding column percentage and computing it using column "Total score" </h3>

```python
percentage = (df_data_1["Total score"]/300)*100
df_data_1["Percentage"]= percentage
```

```python
df_data_1.head()
```

<h3> Computing grade by applying condition to percentage </h3>

```python
def grades(percentage):
     if percentage>85 and percentage<=100:
        return 'Grade A'
     if percentage>70 and percentage<=85:
        return 'Grade B'
     if percentage>55 and percentage<=70:
        return 'Grade C'
     if percentage>33 and percentage<=55:
        return 'Grade D'
     if percentage>0 and percentage<=33:
        return 'Grade E'
df_data_1['Grades']=df_data_1['Percentage'].apply(grades)    
```

```python
df_data_1.head()
```
